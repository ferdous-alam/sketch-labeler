```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('midsection_no_text.png')  # Replace with your image path

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Canny edge detection
edges = cv2.Canny(gray, threshold1=10, threshold2=100)

# Display results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Grayscale Image')
plt.imshow(gray, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Canny Edges')
plt.imshow(edges, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image and apply Canny
img = cv2.imread('midsection_no_text.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 100, 200)

# Get coordinates of edge pixels
y_coords, x_coords = np.where(edges > 0)

# Optionally sample N points
N = 10000
if len(x_coords) > N:
    idx = np.random.choice(len(x_coords), N, replace=False)
    x_coords = x_coords[idx]
    y_coords = y_coords[idx]

# Plot the sampled edge points
plt.figure(figsize=(6, 6))
plt.scatter(x_coords, y_coords, s=1, color='black')
plt.gca().invert_yaxis()
plt.axis('equal')
plt.title('Sampled Edge Points')
plt.show()


```

```python
from skimage import io, color, filters, morphology
import matplotlib.pyplot as plt

# Load image and preprocess
image = io.imread('midsection_no_text.png', as_gray=True)
binary = image < filters.threshold_otsu(image)  # Invert if needed

# Skeletonize
skeleton = morphology.skeletonize(binary)

# Display
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Binary')
plt.imshow(binary, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Skeletonized')
plt.imshow(skeleton, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('abs_test_1.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB for display

# Define bounding boxes: format (x, y, width, height)
# Replace these with your coordinates
boxes = [
    (50, 30, 100, 60),   # example box
    (200, 150, 80, 120),
]

# Draw bounding boxes
for (x, y, w, h) in boxes:
    cv2.rectangle(img, (x, y), (x + w, y + h), color=(255, 0, 0), thickness=2)

# Show image
plt.figure(figsize=(10, 6))
plt.imshow(img)
plt.axis('off')
plt.show()
```
